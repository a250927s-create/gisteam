```python
# ===== 初期設定 =====
import pandas as pd
import folium

# ===== 地図の作成 =====
# 日本中心（好きに変えてOK）
m = folium.Map(
    location=[35.6895, 139.6917],  # 東京
    zoom_start=5
)

# ===== サンプルデータ =====
data = pd.DataFrame({
    "name": ["Tokyo", "Osaka", "Fukuoka"],
    "lat": [35.6895, 34.6937, 33.5902],
    "lon": [139.6917, 135.5023, 130.4017]
})

# ===== マーカー追加 =====
for _, row in data.iterrows():
    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=row["name"]
    ).add_to(m)

# ===== 保存 =====
m.save("map.html")

print("地図を map.html に保存しました")
```

```python
import pandas as pd

data = pd.DataFrame({
    "country": ["中国", "ベトナム", "韓国", "フィリピン", "米国", "インドネシア", "ネパール", "インド", "タイ", "台湾"],
    "value": [375308, 122359, 51900, 51207, 42702, 34587, 25136, 24193, 19203, 14019],
    "lat": [35, 21.028472, 37.57, 14.59, 51.3028, 6.12, 27.42, 28.35, 13.75, 23.46],
    "lon": [105, 105.854167, 126.97, 120.98, 0.0741, 106.5, 85.19, 77.12, 100.5, 121]
})

print(data)
```

```python
import folium
from folium.plugins import HeatMap

m = folium.Map(location=[20, 100], zoom_start=3)

heat_data = data[["lat", "lon", "value"]].values.tolist()

HeatMap(heat_data).add_to(m)

m.save("heatmap.html")
```
